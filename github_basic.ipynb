### Branching
- Though revert and reset lets you get back to the previous version, it may still hamper the stable code, especially when many people are collborating.
- Hence, branching
- Branching creates a new branch with all the codes in the master. One may then edit or add new code to th ebranch and verify if it works.
- After verifying the branch may be added to the master branch.
- `git branch` lets you see the current branch
- Creating new branch
- `git checkout -b <branch name>` this command will create a new branch with the given name. "b" stands for branch
    - in this eg we shall create a new branch named model using code `git checkout -b model`
- runnig this command will create the branch and switch to the branch.
- run `git branch` again and you will see two branches master and model. The * infront of model   indicates that the branch we are working on and any new commit will go into that branch.
- HAving switched the branch lets go to VS Code and add new code to the python code.
    - go to VS Code and create a new file model
- Pusihng the model
    - Note that when you push the code after commit, use the name of the branch currently in use and not the master
    - git push origin <model>
- Go to github and clik on the tab with master. inside you can see a new branch model has been added.
- **Note**
    - git push -u will push all the untracked files to the branch where as git push will add only the current file
    - git push origin will add all the files to all the branches and git push origin <branch name> will add files to the particular branch.
- Switch branches
    - use command `git checkout <branch name>` this will switch to the desired branch. 
    - In case you entered a wrong branch name, it will display an error message.
    - **note** using `git checkout -b <branch name >` creates a new branch
