## Git Basic commands
- First create a new repository in Github
    - Go to github
    - Create New 
    - For the time being leave all options as default and make it public.
    - `Click Create`
    - `Clicking Create` will take you to a page with basic git commands to help comit the file.
- Create a directory using the terminal
    - use `ls` command to display all the folders in the current directory.
    - use `cd <directory name >to enter into any dircetory.
        - use the entire path insted of single directory name if you know it.
        - **note** 
            - if the directory names contains spaces, terminal commands may not recognise it.
            - use`my_directory' instead of `my directory'
    - create new dir using command `mkdir <dir name>
    - Enter the new dir created using `cd <dir name>'
    - **Follow the instructions in the github page**
        - first comd is `echo`
            - syntax is echo "# Github_Basics" >> README.md
                - What ever is in the "" will get written to a file named README.md
            - here we are creating a simple Readme file with a single line in it.
            - use `ls` and see that README.md is displayed
            - `cat <filename>` in this case `cat <README.md>`
                - cat command wwill display what is in the file
         - Initilising a Git repository for the forlder which is created
            - use command `git init`
            - this will initiate an empty repository for the folder and will track all the changes and files in it.
         - Configuring git 
            - This is a one time measure and need not be repeated here after.
            - This is done to connect the local machine with github.
            - Configuration is done based on email and username. For this we use the following commands
                - `git config user.name <your github username>`
                -  `git config user.email <your username liked to github profile>`
                - It will ask for password then use the token as passwords are now obselete in git.

            - Check if the config is working use
                - `git config user.name` & `git config user.email` and it will show you the user name and email
        - Adding new file
            - before adding check the status using `git status`
            - At this stage this command will show the README.md file in read as it is untracked.
            - it will also display 'No commits yet' , 'Nothing added to commit but untracked files present'
            - `git add <file name>` to add the file to staging and then check status again.
                - Adding file will take it to tracking mode from untrack mode.
                - Status will show the file in green, bust will still show no commits yet
        - Commiting the added file
            - use `git comit -m "<message>"
            - -m stands for message and whatever is written in " " followed by -m will be displayed along with the commited file.
            - messages are useful while colaborating or looking at the files later to see what was it about.
            - Press enter and the message will be shown along with 1 file change/// 1 insertion. This means we have added one new file
            - use `git status' and we see the display saying nothing to commit.

        - Renaming the branch
            - use `git branch -M <new name>`
            - This will change the name of the branch to  'new name' you wish to give.
            - for eg you cna change default 'master' to 'main'
            - `git branch` will display the current branch you are working on.
        - Linking repository and local machine folder.
            - Before we can push the files to git repository we need to link th efolder in th elocal machine and the repository
            - Please note that we did the config and that was a one time thing.
            - But for every repository we need to link the repository url to the folder in the local machine.
            - For this go to the github and copy the url, then use following command
            - `git remote add origin <url>`
        
        - Pushing the file to the repository
            - use `git -u push origin <branch name>' 
            - At this stage the system will ask for user name and password
            - type in the username and passowrd prompt will be shown
            - Getting token.
                - PAsswords ar eno more permitted in git, so we need access token.
                - Go to github page and click on the photo in the upper right hand corner.
                - go to <settings>
                - scroll down to <developer settings>
                - go to <personal access token>
                - Write a note to enable further refrence or use of token
                - click on <generate new token>
                - This will show lot of options which you may select as per requirement.
                - scroll down to the end and click <generate>
                - A new token is genrated
                - copy the token
             - paste the token in the password prompt at terminal
                - note that the password is not displayed and hence repasting it will be treated as a wrong token.   
            - it will show master -> master which means its writng from amster in local machine to master in repository.
        - Go to github page and click on the repository
            - you can see the new README.md file created
